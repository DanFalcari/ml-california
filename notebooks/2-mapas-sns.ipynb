```python
# Criando um gráfico com a longitude e latitude somente para observar a distribuição dos dados 
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=df, x="longitude", y="latitude", ax=ax, alpha=SCATTER_ALPHA)

plt.show()
```

```python
# Criando um uma figura com um histograma para observar as regiões com mais concentrações de registros (dados) 
g = sns.jointplot(
    data=df,
    x="longitude",
    y="latitude",
    alpha=SCATTER_ALPHA,
    height=8,
    ratio=4,
)

# Adicionando título e ajustes estéticos
g.fig.suptitle('Distribuição Geográfica das Propriedades', 
               y=1.02, fontsize=14)

# Personalizando os eixos
g.ax_joint.set_xlabel('Longitude', fontsize=12)
g.ax_joint.set_ylabel('Latitude', fontsize=12)

# Adicionando grid para melhor referência
g.ax_joint.grid(True, linestyle='--', alpha=0.3)


plt.tight_layout()
plt.show()
```

### Distribuição Espacial
Foi criado um gráfico de dispersão para visualizar a distribuição geográfica dos dados, confirmando o formato do território analisado
Histogramas complementares revelaram maior concentração de registros em determinadas regiões da Califórnia


```python
# Configurações do gráfico
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(
    data=df,
    x="longitude",
    y="latitude",
    hue="ocean_proximity",
    alpha=SCATTER_ALPHA,        
)

# Título e rótulos
plt.title('Proximidade com o Oceano por Coordenadas Geográficas', fontsize=16, pad=20)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

# Legenda melhorada
legend = ax.legend(
    title='Proximidade do Oceano',
    title_fontsize='13',
    fontsize='11',
    frameon=True,
    framealpha=0.8,
    loc='upper right'
)

# Ajustar os nomes da legenda (se necessário)
new_labels = {
    'NEAR BAY': 'Próximo à Baía',
    '<1H OCEAN': 'Menos de 1h do Oceano',
    'INLAND': 'Interior',
    'NEAR OCEAN': 'Próximo ao Oceano',
}
for text in legend.get_texts():
    text.set_text(new_labels.get(text.get_text(), text.get_text()))

# Grade e estilo
sns.set_style("whitegrid")
plt.grid(True, alpha=0.3)

# Ajustar layout
plt.tight_layout()
plt.show()
```

### Categorização por Proximidade do Oceana
 
O gráfico mostra como as regiões categóricas estão distrbuidas no mapa da Califórnia.{

```python
# Configurações do gráfico
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(
    data=df,
    x="longitude",
    y="latitude",
    hue="median_income_cat",
    palette=PALETTE,  
)

# Título e rótulos
plt.title('Distribuição Geográfica por Renda Mediana Anual (Categórica)', fontsize=16, pad=20)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

# Legenda melhorada
legend = ax.legend(
    title='Renda mediana anual',
    title_fontsize='13',
    fontsize='11',
    frameon=True,
    framealpha=0.8,
    loc='upper right'
)

# Dicionário de mapeamento correto
new_labels = {
    1: "Até $15.000",
    2: "$15.000 - $30.000",
    3: "$30.000 - $45.000",
    4: "$45.000 - $60.000", 
    5: "Acima de $60.000"
}

# Aplicando às legendas existentes
legend = ax.get_legend()
for text in legend.get_texts():
    original_text = text.get_text()
    # Converte texto para inteiro (se necessário)
    try:
        key = int(original_text)
        text.set_text(new_labels.get(key, original_text))
    except ValueError:
        text.set_text(original_text) 

# Grade e estilo
sns.set_style("whitegrid")
plt.grid(True, alpha=0.3)

# Ajustar layout
plt.tight_layout()
plt.show()
```


### Distribuição de Renda
Maiores rendas estão mais concentradas próximas ao litoral e a báia de São Francisco, mostrando uma redução gradual da renda em direção ao interior



```python
# Configurações do gráfico
plt.figure(figsize=(10, 6))
ax = plt.gca()

# Normalização dos valores para a barra de cores
norm = plt.Normalize(df["median_house_value"].min(), df["median_house_value"].max())

# Criando o scatter plot
scatter = sns.scatterplot(
    data=df,
    x="longitude",
    y="latitude",
    hue="median_house_value",
    palette=PALETTE,
    edgecolor='none', 
    hue_norm=(df["median_house_value"].min(), df["median_house_value"].max()),
    ax=ax,
)

# Removendo a legenda automática
ax.get_legend().remove()

# Adicionando barra de cores profissional
cbar = plt.colorbar(
    plt.cm.ScalarMappable(norm=norm, cmap=PALETTE),
    ax=ax,
    orientation='vertical',
    shrink=0.90,  # Tamanho da barra
    label='Valor da Mediana das Casas (US$)',  # Rótulo descritivo
)
# Ajuste a fonte do rótulo
cbar.ax.yaxis.label.set_size(10)

# Ajustes estéticos
ax.set_title('Distribuição Geográfica do Valor da Mediana das Propriedades na Califórnia', pad=20, fontsize=14)
ax.set_xlabel('Longitude', fontsize=10)
ax.set_ylabel('Latitude', fontsize=10)
ax.grid(True, linestyle='--', alpha=0.3)  # Grade suave

# Melhorando o layout
plt.tight_layout()
plt.show()
```
